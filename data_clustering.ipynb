# Import packages

```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

# Read the dataset

```python
df_train = pd.read_csv("./dataset/fraudTrain.csv")
df_train
```

```python
df_test = pd.read_csv("./dataset/fraudTest.csv")
df_test
```

```python
df = pd.concat([df_train, df_test])
df
```

# Cluster it and save the dataset

```python
cluster_df = df[df['is_fraud'] == 1].copy()
cluster_df['trans_date_trans_time'] = pd.to_datetime(cluster_df['trans_date_trans_time']).apply(lambda x: int(x.timestamp()))
cluster_df['dob'] = pd.to_datetime(cluster_df['dob'])
cluster_df.drop(['cc_num', 'merchant', 'first', 'last', 'street', 'zip', 'trans_num', 'dob', 'unix_time', 'is_fraud', 'Unnamed: 0'], axis=1, inplace=True)
cluster_df
```

```python
category_le = LabelEncoder()
gender_le = LabelEncoder()
city_le = LabelEncoder()
state_le = LabelEncoder()
job_le = LabelEncoder()
cluster_df['category'] = category_le.fit_transform(cluster_df['category'])
cluster_df['gender'] = gender_le.fit_transform(cluster_df['gender'])
cluster_df['city'] = city_le.fit_transform(cluster_df['city'])
cluster_df['state'] = state_le.fit_transform(cluster_df['state'])
cluster_df['job'] = job_le.fit_transform(cluster_df['job'])
```

```python
cluster_df
```

```python
cluster_df.info()
```

```python
# 0 - Not fraud
# 1 - Phishing
# 2 - Skimming
# 3 - Snooping
kmeans = KMeans(n_clusters=4 - 1, random_state=42)
cluster_df['Cluster'] = kmeans.fit_predict(cluster_df)
```

```python
cluster_df
```

```python
def apply_cluster(x):
    if x['is_fraud'] == 0:
        return "Not fraud"
    
    x['trans_date_trans_time'] = int(datetime.strptime(x['trans_date_trans_time'], '%Y-%m-%d %H:%M:%S').timestamp())
    x = x.drop(['cc_num', 'merchant', 'first', 'last', 'street', 'zip', 'trans_num', 'dob', 'unix_time', 'is_fraud', 'Unnamed: 0'])

    # Apply the transformations for categorical features
    x['category'] = category_le.transform([x['category']])[0]
    x['gender'] = gender_le.transform([x['gender']])[0]
    x['city'] = city_le.transform([x['city']])[0]
    x['state'] = state_le.transform([x['state']])[0]
    x['job'] = job_le.transform([x['job']])[0]

    x = pd.DataFrame([x])
    cluster = kmeans.predict(x)
    if cluster == 0:
        return "Phishing"
    elif cluster == 1:
        return "Skimming"
    else:
        return "Snooping"

df['fraud'] = df.apply(apply_cluster, axis=1)
df
```

```python
df.to_csv('dataset/final.csv', index=False)
```
